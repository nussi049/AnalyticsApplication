```python
import googlemaps
import pandas as pd

# API with google maps cloud plattform 200$ free contingent.
api_key = ""
```

```python
df_bikes = pd.read_csv("data/sf_2018.csv")
df_bikes
```

```python
# union of two sets without duplicates
stations = list(set(df_bikes["start_station_name"].dropna()) | set(df_bikes["end_station_name"].dropna()))
print(len(stations))
```

```python
geo_df = pd.DataFrame(columns=["station", "lat", "lng", "result"])

# Create googlemaps Client
maps_client = googlemaps.Client(key=api_key)

for station in stations:
    if station != 'nan':
        # Be sure that this street is found in San Francisco
        station = station + ", San Francisco"
        
        # Get geo data from google api
        result = maps_client.geocode(station)

        location = result[0]['geometry']['location']

        data_point = {"station": station, "lat": location["lat"], "lng": location["lng"], "result": result}
        geo_df = geo_df.append(data_point, ignore_index=True)
```

```python
geo_df.to_csv("geo_data.csv", index=False)
```
