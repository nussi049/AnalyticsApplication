```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.tseries.holiday import USFederalHolidayCalendar
from datetime import date
```

```python
df_bikes = pd.read_csv("data/sf_2018.csv")
df_bikes
```

```python
# Drop all NaN values
df_bikes=df_bikes.dropna().reset_index(drop=True)
df_bikes
```

```python
df_weather = pd.read_csv("data/weather_hourly_sf.csv")
df_weather
```

```python
# Drop all NaN values
df_weather=df_weather.dropna().reset_index(drop=True)
df_weather
```

```python
# Split datetime into year, month, day, day_name and hour columns for both dataframes
df_bikes['year'] = pd.to_datetime(df_bikes['start_time']).dt.year
df_bikes['month'] = pd.to_datetime(df_bikes['start_time']).dt.month
df_bikes['day'] = pd.to_datetime(df_bikes['start_time']).dt.day
df_bikes['hour'] = pd.to_datetime(df_bikes['start_time']).dt.hour
df_bikes['day_name'] = pd.to_datetime(df_bikes['start_time'], format='%Y-%m-%d %H:%M:%S').dt.day_name()


df_weather['year'] = pd.to_datetime(df_weather['date_time']).dt.year
df_weather['month'] = pd.to_datetime(df_weather['date_time']).dt.month
df_weather['day'] = pd.to_datetime(df_weather['date_time']).dt.day
df_weather['hour'] = pd.to_datetime(df_weather['date_time']).dt.hour

df_bikes
```

```python
df_weather
```

```python
# Our model should just predict the demand of an hour.
# So we drop the minutes an second value of the bikes dataframe.
# The we merge the temperature and bikes data together, where date_time = start_time
df_merged = pd.merge(df_bikes, df_weather,  how='left', left_on=['year','month', 'day', 'hour'], right_on = ['year','month', 'day', 'hour'])
df_merged = df_merged.dropna().reset_index(drop=True)
df_merged
```

```python
# Now select relevant columns e.g Features for prediction model and save the dataframes.
# For prediction we are using the month, day, hour, max_temp, min_temp and precip of the beginnig hour.
# (We are taking also the start_station_id if we want to predict the demand on a specific station at a specific time)
# We assume that our model should predict the demand of an hour at the beginning of the hour, 
# so that we now the temp and precip values of this hour.

df_filtered = df_merged.filter(['day_name', 'month','day','hour', 'max_temp', 'min_temp', 'precip', 'user_type'], axis=1)
df_filtered
```

```python
# aggregate to get the target column.
df_aggregated_demand = df_filtered.groupby(['day_name','month', 'day', 'hour', 'max_temp', 'min_temp', 'precip']).user_type.agg('count').to_frame('demand').reset_index()
df_aggregated_demand
```

```python
# Now split dataframe into subscribers and customers frame, because both have other patterns for demanding a bike.
# So we will later build to models. One for predicting customers and one for predicting subscribers demand.
# Both models will predict a demand which will be summed up.
df_subscriber = df_filtered[df_filtered['user_type'] == "Subscriber"]
df_customer = df_filtered[df_filtered['user_type'] == "Customer"]

df_subscriber
```

```python
df_customer
```

```python
# Now we have to aggregate to get the target column.
agg_subscriber = df_subscriber.groupby(['day_name','month', 'day', 'hour', 'max_temp', 'min_temp', 'precip']).user_type.agg('count').to_frame('demand').reset_index()
agg_subscriber
```

```python
# Do the same for the customer frames
agg_customer = df_customer.groupby(['day_name', 'month', 'day', 'hour', 'max_temp', 'min_temp', 'precip']).user_type.agg('count').to_frame('demand').reset_index()
agg_customer
```

```python
# Now save all three dataframes
df_aggregated_demand.to_csv("data/all_demand.csv", index=False)

agg_subscriber.to_csv("data/subscriber_demand.csv", index=False)

agg_customer.to_csv("data/customer_demand.csv", index=False)
```
